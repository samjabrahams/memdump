```python
%matplotlib inline
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
```

# Visualize Event Data (from Santa Monica API)

```python
#Meter 1 Event Times
x1_s = [4,8]
x1_e = [2,5,10]

#Meter 2 Event Times
x2_s = [1,9]
x2_e = [7]
```

```python
#Meter 1 Start Events
s1 = [1,1]
#Meter 1 End Events
e1 = [1,1,1]

#Meter 2 Start Events
s2 = [2,2]
#Meter 2 End Events
e2 = [2]
```

```python
plt.figure(figsize=(8,3))
event1 = plt.scatter(x1_s,s1, s=50, label='Meter 1 Event')
plt.scatter(x1_e,e1, s=50)

event2 = plt.scatter(x2_s,s2, s=50, label='Meter 2 Event', color='r')
plt.scatter(x2_e,e2, s=50, color='r')

plt.xlabel('Time')
plt.ylabel('Meter ID')

plt.legend(handles=[event1, event2],
           scatterpoints=1, loc='center right')
plt.title('Event Data (Before transformation)')
```

# Assign Event Type To Data

```python
plt.figure(figsize=(8,3))
start = plt.scatter(x1_s,s1, marker='v', c='green', s=150)
end = plt.scatter(x1_e,e1, marker='^', c='red', s=150)

plt.scatter(x2_s,s2, marker='v', c='green', s=150)
plt.scatter(x2_e,e2, marker='^', c='red', s=150)

plt.xlabel('Time')
plt.ylabel('Meter ID')

plt.legend((start, end), ('Start Event', 'End Event'),
           scatterpoints=1, loc='center right')
plt.title('Event Data (Event Type Assigned)')
```

# Connecting Events to Form Sessions

```python
#Meter 1 connected events
event1_1 = [-1,2]
event1_2 = [4,5]
event1_3 = [8,10]

#meter 2 connected events
event2_1 = [1,7]
event2_2 = [9,13]

meter_1 = [1,1]
meter_2 = [2,2]
```

```python
plt.figure(figsize=(8,3))
m1 = plt.plot(event1_1, meter_1, '-bo', label='Meter 1 Session')
plt.plot(event1_2, meter_1, '-bo')
plt.plot(event1_3, meter_1, '-bo')

m2 = plt.plot(event2_1, meter_2, '-ro', label='Meter 2 Session')
plt.plot(event2_2, meter_2, '-ro')

plt.text(0.6,1.1,'Session 8', fontsize=8)
plt.text(4,1.1,'Session 10', fontsize=8)
plt.text(8.5,1.1,'Session 11', fontsize=8)

plt.text(3.5,2.1,'Session 9', fontsize=8)
plt.text(10,2.1,'Session 12', fontsize=8)

plt.xlabel('Time')
plt.ylabel('Meter ID')

plt.title('Events Converted to Sessions')
plt.axis([0,12,0,3.5])
plt.legend()
```

# Using Session Cross-Sections to Make Balanced Panel Data

```python
#Meters Available
a1 = np.concatenate((np.arange(2,4,0.5), np.arange(5,8,0.5),np.arange(10,13,0.5)))
a2 = np.concatenate((np.arange(0,1,0.5), np.arange(7,9,0.5)))

#Meters Occupied
o1 = np.concatenate((np.arange(0,2,0.5),np.arange(4,5,0.5),np.arange(8,10,0.5)))
o2 = np.concatenate((np.arange(1,7,0.5),np.arange(9,13,0.5)))
```

```python
meter1_available = np.empty([1,16])
meter2_available = np.empty([1,6])

meter1_occupied = np.empty([1,10])
meter2_occupied = np.empty([1,20])

meter1_available.fill(1)
meter1_occupied.fill(1)
meter2_available.fill(2)
meter2_occupied.fill(2)
```

```python
plt.figure(figsize=(8,3))

free = plt.scatter(a1, meter1_available, c='green', marker='o')
occ = plt.scatter(o1, meter1_occupied, c='red', marker='x')

plt.scatter(a2, meter2_available, c='green', marker='o')
plt.scatter(o2, meter2_occupied, c='red', marker='x')

plt.xlabel('Time')
plt.ylabel('Meter ID')

plt.axis([0,12,0,3.5])
plt.legend((free,occ), ('Space Availble', 'Space Occupied'), scatterpoints=1)
plt.title('Balanced Panel Data')
```
